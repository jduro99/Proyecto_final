```python
# --{IMPORT OF NECESSARY LIBRARIES}--
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px


dataset = pd.read_csv("D:\eljl2\Documents\Repositorios Github\Finanzas_personales_India\data\Finanzas_India.csv")
```

```python
# --{NUMBER OF DUPLICATE ROWS AND NULLS PER ROW}--

# Count the duplicate rows in dataset:
df_duplicate_rows_dataset = dataset.duplicated()
print('Total number of duplicate rows in dataset: ', dataset.duplicated().sum())
print(df_duplicate_rows_dataset)

# Count the null values per row in dataset:
dataset['nulls_per_row'] = dataset.isnull().sum(axis=1)
print(dataset['nulls_per_row'])
```

```python
dataset = dataset.round(2)
print(dataset)
```

```python
# Agrupar las edades en categorías
bins = [0, 20, 30, 40, 50, 60, 70, 80, 90, 100]
labels = ['0-20', '21-30', '31-40', '41-50', '51-60', '61-70', '71-80', '81-90', '91-100']
dataset['Age_Group'] = pd.cut(dataset['Age'], bins=bins, labels=labels, right=False)

# Crear el gráfico de violín con la agrupación de edades
sns.violinplot(x='Loan_Repayment', y='Age_Group', data=dataset)
plt.title('Violin Plot of Loan Repayment Distribution by Age Group')
plt.xlabel('Loan Repayment')
plt.ylabel('Age Group')
plt.show()
```

Hay demasiados otuliers en loan_payment, crearemos grupos para hacer un repressentacion mas clara de la como esta la situcion de prestamos en base a la franja de edad de la poblacion.

```python
# Agrupar las edades en categorías
bins_age = [0, 20, 30, 40, 50, 60, 70, 80, 90, 100]
labels_age = ['0-20', '21-30', '31-40', '41-50', '51-60', '61-70', '71-80', '81-90', '91-100']
dataset['Age_Group'] = pd.cut(dataset['Age'], bins=bins_age, labels=labels_age, right=False)

# Agrupar los valores de Loan_Repayment en categorías
bins_loan = [-1, 1000, 5000, 10000, 20000, 50000, 100000]
labels_loan = ['0-1k', '1k-5k', '5k-10k', '10k-20k', '20k-50k', '50k-100k']
dataset['Loan_Group'] = pd.cut(dataset['Loan_Repayment'], bins=bins_loan, labels=labels_loan, right=False)

print(dataset[['Age', 'Age_Group', 'Loan_Repayment', 'Loan_Group']])
```

```python
# Crear una nueva tabla con las columnas Age, Age_Group y Loan_Group
new_table = dataset[['Age', 'Age_Group', 'Loan_Group']]

# Calcular los cuartiles y el IQR para la columna Loan_Repayment
Q1_loan = dataset['Loan_Repayment'].quantile(0.25)
Q3_loan = dataset['Loan_Repayment'].quantile(0.75)
IQR_loan = Q3_loan - Q1_loan

# Definir los límites inferior y superior para la limpieza de outliers
lower_bound_loan = Q1_loan - 1.5 * IQR_loan
upper_bound_loan = Q3_loan + 1.5 * IQR_loan

# Aplicar la limpieza de outliers en la columna Loan_Repayment
new_table_cleaned = new_table[(dataset['Loan_Repayment'] >= lower_bound_loan) & (dataset['Loan_Repayment'] <= upper_bound_loan)]

print(new_table_cleaned)
```

```python
# Crear el gráfico de violín con la agrupación de edades
sns.violinplot(x='Loan_Group', y='Age_Group', data=new_table_cleaned)
plt.title('Violin Plot of Loan Group Distribution by Age Group')
plt.xlabel('Loan Group')
plt.ylabel('Age Group')
plt.show()
```
