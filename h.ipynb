```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df=pd.read_csv('datos_modelo.csv')
df
```

```python
data=pd.read_csv('data/data.csv')
data
```

```python
import plotly.graph_objects as go
import pandas as pd
 
# Agrupar los datos por City_Tier y Occupation y calcular la suma de Gastos_Fijos y Gastos_Variables
gastos_por_city_tier_occupation = df.groupby(['City_Tier', 'Occupation'])[['Gasto_Fijo', 'Gastos_variables']].sum().reset_index()
 
# Crear una lista con las combinaciones de City_Tier y Occupation como etiquetas
labels = gastos_por_city_tier_occupation['City_Tier'] + " - " + gastos_por_city_tier_occupation['Occupation']
 
# Crear el gráfico de barras apiladas con fondo negro
fig = go.Figure()
 
# Añadir barras para Gastos Fijos
fig.add_trace(go.Bar(
    x=labels,
    y=gastos_por_city_tier_occupation['Gasto_Fijo'],
    name='Gasto Fijo',
    marker_color='#8B0000'  # Rojo oscuro
))
 
# Añadir barras para Gastos Variables
fig.add_trace(go.Bar(
    x=labels,
    y=gastos_por_city_tier_occupation['Gastos_variables'],
    name='Gasto Variable',
    marker_color='#00008B'  # Azul oscuro
))
 
# Personalizar el diseño del gráfico
fig.update_layout(
    barmode='stack',  # Barras apiladas
    title='Gastos Fijos y Variables por City_Tier y Occupation',
    xaxis_title='City_Tier - Occupation',
    yaxis_title='Total Gastos',
    template='plotly_dark',  # Tema oscuro
    width=1000,  # Ancho del gráfico
    height=600,  # Altura del gráfico
    legend=dict(
        title='Tipo de Gasto',
        orientation="h",  # Leyenda horizontal
        yanchor="bottom",
        y=1.02,
        xanchor="center",
        x=0.5
    )
)
 
# Ajustar las etiquetas del eje X
fig.update_xaxes(tickangle=45)
 
# Mostrar el gráfico
fig.show()
```

```python
import plotly.graph_objects as go
import pandas as pd
 
# Agrupar ingresos promedio por edad
income_by_age = df.groupby('Age')['Income'].mean()
overall_mean_income = df['Income'].mean()
 
# Crear la figura con Plotly
fig = go.Figure()
 
# Agregar la línea de ingresos promedio por edad
fig.add_trace(go.Scatter(
    x=income_by_age.index,
    y=income_by_age.values,
    mode='lines+markers',
    name='Ingreso promedio por edad',
    line=dict(color='blue'),
    marker=dict(size=8)
))
 
# Agregar la línea horizontal para el ingreso promedio general
fig.add_trace(go.Scatter(
    x=income_by_age.index,
    y=[overall_mean_income] * len(income_by_age),
    mode='lines',
    name='Ingreso promedio general',
    line=dict(color='red', dash='dash')
))
 
# Personalizar el diseño del gráfico
fig.update_layout(
    title='Promedio de Ingresos por Edad',
    xaxis_title='Edad',
    yaxis_title='Ingreso Promedio',
    template='plotly_white',  # Tema blanco
    width=1000,  # Ancho del gráfico
    height=600,  # Altura del gráfico
    legend=dict(
        orientation="h",  # Leyenda horizontal
        yanchor="bottom",
        y=1.02,
        xanchor="right",
        x=1
    )
)
 
# Añadir rejilla y estilo del fondo
fig.update_xaxes(showgrid=True, gridwidth=1, gridcolor='lightgray')
fig.update_yaxes(showgrid=True, gridwidth=1, gridcolor='lightgray')
 
# Mostrar el gráfico
fig.show()
```

```python
import plotly.express as px
import pandas as pd
 
# Crear nuevas columnas para el ahorro potencial por categoría
categories = ['Groceries', 'Transport', 'Eating_Out', 'Entertainment', 'Utilities', 'Healthcare', 'Education', 'Miscellaneous']
for category in categories:
    data[f'{category}_savings'] = data['Disposable_Income'] - data[category]
 
# Calcular el ahorro potencial total
data['total_savings'] = data[[f'{category}_savings' for category in categories]].sum(axis=1)
 
# Calcular el promedio de ahorro potencial por categoría
category_savings = {category: data[f'{category}_savings'].mean() for category in categories}
 
# Crear un DataFrame para usar en el gráfico
category_savings_datos = pd.DataFrame(list(category_savings.items()), columns=['Categoría', 'Ahorro Potencial Promedio'])
 
# Crear el gráfico de barras con colores oscuros
fig = px.bar(
    category_savings_datos,
    x='Categoría',
    y='Ahorro Potencial Promedio',
    color='Categoría',
    text='Ahorro Potencial Promedio',
    color_discrete_sequence=px.colors.qualitative.Dark2,  # Paleta de colores oscuros
    title='Promedio de Ahorro Potencial por Categoría'
)
 
# Ajustar el diseño y dimensiones
fig.update_traces(
    texttemplate='%{text:.2f}',  # Mostrar solo 2 decimales
    textposition='outside'
)
fig.update_layout(
    xaxis_title='Categoría de Gasto',
    yaxis_title='Ahorro Potencial Promedio',
    title_font_size=16,
    xaxis_tickangle=45,
    template='plotly_dark',  # Tema oscuro
    width=1000,  # Ancho del gráfico
    height=600   # Altura del gráfico
)
 
# Mostrar el gráfico
fig.show()
```

```python
import plotly.express as px
import pandas as pd
 
# Crear columnas de porcentaje para cada categoría de gasto
categories = ['Groceries', 'Transport', 'Eating_Out', 'Entertainment', 'Utilities', 'Healthcare', 'Education', 'Miscellaneous']
for category in categories:
    data[f'{category}_percentage'] = (data[category] / data['Income']) * 100
 
# Calcular los porcentajes promedio por categoría
category_percentages = [f'{category}_percentage' for category in categories]
mean_percentages = data[category_percentages].mean()
 
# Crear un DataFrame para usar en el gráfico
df_plot = pd.DataFrame({
    'Categoría de Gasto': categories,
    'Porcentaje Promedio': mean_percentages.values
})
 
# Crear el gráfico de barras con colores oscuros
fig = px.bar(
    df_plot,
    x='Categoría de Gasto',
    y='Porcentaje Promedio',
    color='Categoría de Gasto',
    text='Porcentaje Promedio',
    color_discrete_sequence=px.colors.qualitative.Dark2,  # Paleta de colores oscuros
    title='Porcentaje Promedio de Gasto por Categoría sobre el Ingreso Total'
)
 
# Ajustar el diseño y dimensiones
fig.update_traces(
    texttemplate='%{text:.2f}%',  # Mostrar solo 2 decimales
    textposition='outside'
)
fig.update_layout(
    xaxis_title='Categoría de Gasto',
    yaxis_title='Porcentaje sobre el Ingreso',
    title_font_size=16,
    xaxis_tickangle=45,
    template='plotly_dark',  # Tema oscuro
    width=1000,  # Ancho del gráfico
    height=600   # Altura del gráfico
)
 
# Mostrar el gráfico
fig.show()
```

```python
import plotly.graph_objects as go

# Crear el gráfico de dispersión
fig = go.Figure()

# Agregar puntos para Gastos Fijos
fig.add_trace(go.Scatter(
    x=df['Income'],
    y=df['Gasto_Fijo'],
    mode='markers',
    name='Gasto Fijo',
    marker=dict(
        color='blue',
        size=8,
        line=dict(width=1, color='darkblue')  # Bordes oscuros para mayor contraste
    )
))

# Agregar puntos para Gastos Variables
fig.add_trace(go.Scatter(
    x=df['Income'],
    y=df['Gastos_variables'],
    mode='markers',
    name='Gasto Variable',
    marker=dict(
        color='orange',
        size=8,
        line=dict(width=1, color='darkorange')  # Bordes oscuros para mayor contraste
    )
))

# Personalizar el diseño del gráfico
fig.update_layout(
    title='Relación entre Income y Gastos (Fijos vs Variables)',
    xaxis_title='Income',
    yaxis_title='Gastos',
    template='plotly_white',  # Tema claro
    width=1000,  # Ancho del gráfico
    height=600,  # Altura del gráfico
    legend=dict(
        orientation="h",
        yanchor="bottom",
        y=1.02,
        xanchor="center",
        x=0.5
    ),
    font=dict(
        family='Arial',
        size=12
    ),
    plot_bgcolor='#f9f9f9',  # Fondo del gráfico claro
    paper_bgcolor='#f9f9f9'  # Fondo del papel claro
)

# Añadir rejilla para facilitar la interpretación visual
fig.update_xaxes(showgrid=True, gridwidth=1, gridcolor='lightgray')
fig.update_yaxes(showgrid=True, gridwidth=1, gridcolor='lightgray')

# Mostrar el gráfico
fig.show()
```
